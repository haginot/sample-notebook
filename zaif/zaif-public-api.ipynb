```python
from zaifapi import ZaifPublicApi
import pandas as pd
from datetime import datetime as dt
```

```python
zaif = ZaifPublicApi()
```

```python
currencies = zaif.currencies('all')
```

```python
currencies_df = pd.DataFrame.from_dict(currencies)
currencies_df.sort_values('id')
```

```python
currency_pairs = zaif.currency_pairs('all')
```

```python
currency_pairs_df = pd.DataFrame.from_dict(currency_pairs)
currency_pairs_df[currency_pairs_df['event_number'] == 0].sort_values('id')
```

```python
trades = zaif.trades('btc_jpy')
```

```python
trades_df = pd.DataFrame.from_dict(trades)
trades_df['date'] = trades_df['date'].apply(lambda x: dt.fromtimestamp(x))
trades_df.sort_values('date')
```
